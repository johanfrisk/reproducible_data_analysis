```python
# Quick utility to embed the videos below
from IPython.display import YouTubeVideo
def embed_video(index, playlist='PLYCpMb24GpOC704uO9svUrihl-HY1tTJJ'):
    return YouTubeVideo('', index=index - 1, list=playlist, width=600, height=350)
```

```python
import pandas as pd
data = pd.read_csv('Freemont.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
%matplotlib inline
data.resample('W').sum().plot()
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')

data.columns = ['West', 'East']

data.resample('W').sum().plot()
```

```python
data['Total'] = data['West'] + data['East']

ax = data.resample('D').sum().rolling(365).sum().plot()
ax.set_ylim(0, None);
```

```python
data.groupby(data.index.time).mean().plot();
```

```python
pivoted = data.pivot_table('Total', index = data.index.time, columns = data.index.date)
pivoted.iloc[:5, :5]
```

```python
pivoted.plot(legend = False, alpha = 0.01)
```
